## 1) Konsep Kunci (super ringkas)
- **Repo**: rumah proyek.  
- **Commit**: snapshot perubahan.  
- **Branch**: jalur pengembangan (umumnya `main`).  
- **Remote**: repo di GitHub.  
- **Push/Pull**: kirim/ambil perubahan.  
- **.gitignore**: daftar file yang tidak ikut versioning.



## 2) Aturan Praktikum (kebiasaan)
- Sinkronisasi dilakukan **minimal 1× setiap sesi**.  
- Commit kecil, sering: setiap ada kemajuan bermakna (fungsi selesai, bug fix, analisis selesai).  
- Hindari unggah data besar & data sensitif.

## 8) Buat / update `.gitignore`
Agar file sampah tidak ikut repo.


```python
# Buat .gitignore sederhana (aman untuk praktikum)
gitignore_text = """__pycache__/
.ipynb_checkpoints/
*.pyc
*.pyo
*.pyd
.DS_Store
"""

with open(".gitignore", "w", encoding="utf-8") as f:
    f.write(gitignore_text)

print("Isi .gitignore:")
!cat .gitignore
```
